## Probabilidade simples


Qual a probabilidade de uma escola ter uma média positiva de vitórias?

```python
probMediaMaior = len(tabela[tabela['win_percentage'] > 0.5]) /len(tabela)
print(f"A probabilidade de se encontrar um time com uma média positiva de vitórias é de: {probMediaMaior:.4%}")
```

Com essa informação adquirida por meio de probabilidade simples, percebe-se que em todas as temporadas da liga é muito mais provável que um time consiga ter mais vitórias que derrotas.

De acordo com o site https://www.sports-reference.com/cbb/conferences/big-12/2022.html a média geral de vitórias e derrotas da conferência na temporada 2021-2022 foi de 0.633.

Qual a probabilidade de uma escola ser primeira rankeada?

```python
probPrimRank = len(tabela[tabela['rank'] == 1]) /len(tabela)
print(f"A probabilidade de uma escola ser primeira rankeada é de: {probPrimRank:.4%}")
```

Com essa informação adquirida por meio de probabilidade simples, a chance de se selecionar uma escola aleatória que tenha sido primeira ranqueada em alguma temporada é baixa já que cada temporada tem 1 primeira ranquada dentre 10 escolas.

Como curiosidade, eu encontrei no site https://www.sports-reference.com/cbb/conferences/big-12/2022.html que nas últimas 10 temporadas, apenas 4 escolas que foram primeiras ranquadas na temporada regular foram campeãs da conferência.

## Probabilidade de união

Qual a probabilidade de uma escola ser primeira ou última rankeada?

```python
primOuUlt = len(tabela[tabela['rank'] == 1]) | len(tabela[tabela['rank'] == 10])
probPrimOuUlt = primOuUlt/len(tabela)
print(f"A probabilidade de uma escola ser primeira ou última rankeada é de: {probPrimOuUlt:.4%}")
```

Com essa informação adquirida por meio de probabilidae de união, a chance de se selecionar uma escola aleatória que tenha sido primeira ou última ranqueada em alguma temporada é baixa já que cada temporada tem 1 primeira e 1 última ranquada dentre 10 escolas.

## Probabilidade de intersecção

Qual a probabilidade de uma escola ser primeira ranquada e ter média maior que 0.5 de vitórias?

```python
mediaPrim = len(tabela[tabela['rank'] == 1]) & len(tabela[tabela['win_percentage'] > 0.5])
probMediaPrim = mediaPrim/len(tabela)
print(f"A probabilidade de uma escola ser primeira ranquada e ter média maior que 0.5 de vitórias é de {probMediaPrim:.4%}")
```

Com essa informação adquirida por meio de probabilidae de intersecção, a chance de se selecionar na tabela uma escola que seja primeira ranqueada e ter média positiva de vitórias é baixa.

## Probabilidade de complemento

Qual a probabilidade de uma escola ter uma média positiva de vitórias e não ser primeira ou última rankeada?

```python
probNaoPrimOuUlt = probMediaMaior - probPrimOuUlt
print(f"A probabilidade de uma escola ter uma média positiva de vitórias e não ser primeira ou última rankeada é de {probNaoPrimOuUlt:.4%}")
```

Com essa informação adquirida por meio de probabilidae de complemeto, a chance de se selecionar uma escola aleatória que tenha média positiva de vitórias mas não ser primeira ou última ranqueada é alta, porém mais baixa do que apenas selecionar uma escola com média positiva de vitórias, indo de aproximadamente 70% para 61,3%.

## Probabilidade de diferença

Qual a probabilidade de uma escola ser primeira ranquada mas não ter 35 ou mais vitórias?

```python
primMenos35 = len(tabela[tabela['rank']==1]) - len(tabela[(tabela['wins']>=35) & (tabela['rank']==1)]) 
probPrimMenos35 = primMenos35/len(tabela)
print(f"A probabilidade de uma escola ser primeira ranquada mas não ter 35 ou mais vitórias é de: {probPrimMenos35:.4%}")
```

Com essa informação adquirida por meio de probabilidade de diferença, a chance de se selecionar uma escola que seja primeira ranqueada mas não tenha no mínimo 35 vitórias é baixa, já que a maioria das campeãs ganha disso para cima.

## Probabilidade condicional

Qual a probabilidade de uma escola ser primeira ranquada sabendo que a sua média de vitórias foi maior que 0.5?

```python
probCond = probMediaPrim/probMediaMaior
print(f"A probabilidade de uma escola ser primeira ranquada sabendo que a sua média de vitórias foi maior que 0.5 é de: {probCond:.4%}")
```

Com essa informação adquirida por meio de probabilidade condicional, a chance de se selecionar uma escola que seja primeira ranqueada dentre as que tiveram média positiva de vitórias é baixa, já que a maioria das escolas tiveram a média positiva.

## Teorema de Bayes

Qual a  probabilidade de ter mais de 0.5 de média de vitórias tal seja primeira ranqueada?

```python
probTqMediaMaior = (probMediaMaior*probCond)/probPrimRank
print(f"A probabilidade de ser primeira ranqueada tal que tenha mais de 0.5 de média de vitórias é de: {probTqMediaMaior:.4%}")
```

Com essa informação adquirida por meio do Teorema de Bayes, a chance de se selecionar uma escola ter média positivas de vitórias dado que a probabilidade dela ser primeira ranqueada já tenha ocorrido é de 30,76% aproximadamente, que comparado com quando a gente tira a probabilidade de intesecção desses 2 eventos ocorrerem é bem maior.

# Distribuição de probabilidade

```python
import numpy as np
import scipy
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt 

size=2
x = np.arange(size)
y = tabela['win_percentage'].values
h = plt.hist(y, density=True)

dist_names = ['expon', 'logistic', 'norm']
for dist_name in dist_names:
  dist = getattr(scipy.stats, dist_name) 
  params = dist.fit(y)
  arg = params[:-2]
  loc = params[-2]
  scale = params[-1]
  if arg: 
    pdf_fitted = dist.pdf(x, *arg, loc=loc, scale=scale)
  else:
    pdf_fitted = dist.pdf(x, loc=loc, scale=scale)
  plt.plot(pdf_fitted, label=dist_name)

plt.legend(loc= 'upper right')
```

Nesse gráfico é possível observar a distribuição de probabilidade de se obter tal porcentagem de vitórias, e como conseguimos ver a maioria das porcentagens estão entre 0.6 e 0.8.

# Distribuição amostral

```python
from numpy.random import seed
from numpy.random import randint
from numpy import mean

seed(1)

df = pd.DataFrame(tabela, columns = ['wins'])
df = df.sample(n = 50)
print(df)
print(mean(df))
```

Essa parte do código foi feita para pegar 50 valores aleatórios, sem reposição, na coluna 'wins' da tabela e fazer a médias destes. 

```python
from matplotlib import pyplot

means = []
for _ in range(1000):
  df = pd.DataFrame(tabela, columns = ['wins'])
  df = df.sample(n = 1)
  means.append(int(mean(df)))
  
pyplot.hist(means)
pyplot.show()
```

Com essa forma de pegar números aleatório, peguei 1000 vezes 1 valor aleatório da mesma coluna e fazendo a distribuição amostral deles foi possível concluir que a maioria dos valores retirados foram entre 15 e 25 vitórias.

```python
means = []
for _ in range(1000):
  df = pd.DataFrame(tabela, columns = ['wins'])
  df = df.sample(n = 100)
  means.append(int(mean(df)))
  
pyplot.hist(means)
pyplot.show()
```

Agora fiz a mesma coisa que no anterior porém pegando 1000 vezes amostras de 100 valores e fazendo as médias deles, com isso fazendo denovo o gráfico da distribuição amostral é possível notar que a grande maioria foi perto do 19 e do 20.

```python
means = []
for _ in range(1000):
  df = pd.DataFrame(tabela, columns = ['wins'])
  df = df.sample(n = 290, replace = True)
  means.append(int(mean(df)))

pyplot.hist(means)
pyplot.show()
```

Fazendo a mesma coisa porém dessa vez com reposição e utilizando todas as linhas da tabela, é possível perceber que mesmo assim não fugiu à regra e se manteve sendo a maioria na mesma média. Também é possível notar que quanto menos amostras mais varia o valor das médias.

# Intervalo de confiança

```python
import statsmodels.formula.api as smf
tabela['winsperlosses'] = tabela['wins']/tabela['losses']
df2 = pd.DataFrame(tabela, columns = ['wins', 'losses', 'winsperlosses'])

modelo = smf.ols('winsperlosses ~ wins + losses', data=df2)
resultado = modelo.fit()
resultado.summary()
```

```python
resultado.conf_int(alpha = 0.05)
```

```python
resultado.conf_int(alpha = 0.1)
```

O calculo do intervalo de confiança é feito para ver a confiaça de que o valor real de 'winsperlosses', esteja entre 'wins' e 'losses', sendo o primeiro de 95% e o segundo de 90%
